Data **type** or simply **type** is an attribute of the data which tells a compiler or interpreter how we intend to use the data.

### The `Nominative` Bit about Julia Types

Julia types are origanized into a **tree** hierarchical structures.


Tree roots are a special type called `Any`, all other types are connected to this `Any` root directly or indirectly. 

The `Any` type has children nodes called **subtypes**.  `Any's` children have all have the same parent **supertype** `Any`.


One other particularly distinctive feature of Julia's type system is the notion of **concrete** types.  These are the leaves of the trees. Leaves have NO **subtypes**. **Concrete** types are the final types along a tree branch. Only concrete types can be instantiated in Julia code. 

As a result, every non-leaf node of a tree is known as an **abstract** type.   All leaf, **concrete** types have **abstract** types as their **supertypes**. **Abstract** types (e.g.,`Any`, `Number`) allow use to write code that generalizes for any of there subtypes. 

While this might at first seem unduly restrictive, it has many beneficial consequences with surprisingly few drawbacks

## Type Assertions and Assignments:  Attaching Type Annotations to Expressions and Variables 

Why? 

Answer- help ensure that Julia programs work as expected and provide extra type information to the compiler which improves performance.


use the operator `::` which is read as ***"is an instance of"*** to attach the type annotations.


### Assertions -- `::` as the middleman

model: **(expression on the left)** `::`  **(`type` on the right) **

 *Asserts* that the value of the *expression* on the left *"is an instance of"* the `type` on the right. 
 
*This technique helps confirm that programs work as expected*


````julia
(1+2)::Int64   # (1+2) is read as "is an instance of" Int64
````

````julia
# Expression not an instance of the type on the right
(1+2)::Float64
````

### Assignments
* `::` on the left side of the assignment (=) operator 
* `::` After a function definition

models: 

**variable** `::` **Specified type** = **value**

  The `variable` is always of the `specified type` and the `value` is converted to the `specified type` if needed. 
    

**functionDefinition** `::` **type** 

   the `return value` of the function is always converted to the `specified type`


````julia
# Variable Assignment example
function returnfloat()
    x::Float64 = 100   #force conversion
    x
end

resultReturnFloat = returnfloat()
display(resultReturnFloat)
display(typeof(resultReturnFloat))
````

````julia
# Function example
function sinsquared(x)::Float64
    if x == 0
        return 1
    end
    sin(x)* sin(x)
end

display(sinsquared(0))
display(sinsquared(π))
````

## Creating User Specific Unique Abstract Types : `Subtyping` 

Kind of like creating your own new tree branch off of the existing Julia Type tree. 

Here, we know any type has `Any` as its **supertype**


Abstract types are declared using the `abstract type` keyword. The `abstract type` keyword introduces a new abstract type. The name can be optionally followed by `<:` an already-existing abstract type, indicating that the newly declared abstract type is a **subtype** of this "parent type"

When no **supertype** is given the default supertype is `Any` - a predefined abstract type that all objects are instances of and all types are subtypes of. 

The general syntax for declaring an abstracttype is:

 * ```abstract type «name» end```
 
 or

* ```abstract type «name» <: «supertype» end```


`<:` is read *"is a subtype of"*




````julia
Integer <: Number  # Integer is a subytype of Number 
````

````julia
Integer <: AbstractFloat # Integer is not a subtype of Abstract Float
````

## We can also build `Composite Types` to go beyond Julia's Built in Types

A programmer-defined `composite` type is also called a *struct*.

``` 
struct <Name>  # Header indicates new struct name
        field1     # body of struct with attribute/field1 
          .
          .
          .
        fieldn     # body of struct with attribute/fieldn
end
 ```

````julia
# Composite type names are capitalized and camel case
# Foo is a mix 3 types: ANY, Int64, and Float64 types
struct Foo
    bar 
    baz::Int64
    qux::Float64
end

foo = Foo("Hello, world.", 23, 1.5)  #to create a Foo call it like a function with arguments the value of the fields


display(typeof(Foo))  # Composite type/ struct  Foo is of DataType
display(typeof(foo))  # type of foo is Composite type Foo
foo # returns a reference to a Foo boject. 
````
