## Vectors
They are a form of array, they are not matrices.

Let's have a look

```julia
# Column vectors (many rows x 1 column)
v = [1 , 2, 3, 4]
display(typeof(v))
display(v)
```

```julia
# Make the same column vector using vertical concatenation with ;'s
vv = [1;2;3;4]
display(typeof(vv))
display(vv)
```

```julia
# Make the same column vector using vertical concatenation vcat
vvv = vcat(1,2,3,4)
display(typeof(vvv))
display(vvv)
```

```julia
#See all the column vectors are equivalent
v==vv && vv==vvv && v==vvv
```

```julia
#Row vectors (1 row x many columns)
w = [1 2 3 4]
display(typeof(w))
display(w)
```

```julia
# Make the same row vector using hcat concatentation
ww = hcat(1, 2, 3, 4)
display(typeof(ww))
display(ww)
```

```julia
#Row vectors are equivalent
w == ww
```

# Loops

## While Loops

The syntax for **while** is

    while *condition*
    
        *loop body*
            
    end
       

```julia
n = 0
while n < 10
    n+= 1
    println(n)
end
```

## For Loops

for *var* in * loop iterable *
 
    *loop body*
end



```julia
for n in 1:10
  println(n)
end
```

we can replace *in* with $\in$ or = 

```julia
#for n  1:10
for n = 1:10
  println(n)
end
```

```julia
#For loops with 2-dimensions loops in loops
m ,n = 5,5
A = zeros(m,n)
```

```julia
for i in 1:m
    for j in 1:n
        A[i,j] = i+j
    end
end
A
```

Alternatively, we can use some **synatic sugar** for the same nested `for` loop. **Syntactic Sugar** means a coding simplification to get simiar result.

```julia
B = zeros(m,n)
for i in 1:m, j in 1:n 
    B[i,j] = i + j
end
B
```

The more "Julia" way to do it is to use an **array comprehension**

```julia
C = [i + j for i in 1:m, j in 1:n] #Builds C without having to do the zeros
```

```julia
display(C)
```

# Execption Handling
When there is an error in the code, we throw an exception

During execution we get a failure.  For real-time applications we do not want the program to collapse

Every type of error has an exception message

use "**try-catch** block" to catch exceptions

```julia
# make exception occur

"string" + 1
```

```julia
# Try catch error catch possible error
try
    sqrt(-5)
catch
    println("An exception has occurred")
end
```

```julia
# Try catch error catch catch the exception objec type
try
    sqrt(-5)
    catch x  #here x is the exception object type
    println("An exception has occurred- ", x)
end
```

```julia
try 
    arr =[4,-5]
    sqrt("string")
catch x
    if isa(x,DomainError)
        println("An exception has occurred- incorrect parameter - ",x)
    elseif isa(x,BoundsError)
        println("An exception has occurred- incorrect array index ",x)
    else
        println("An exception has occurred - ", x)
    end
end
```
